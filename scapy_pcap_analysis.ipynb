```python
#!/usr/bin/python
import sys
from scapy.all import *
```

```python
#Get capture file and drive path from user input
capturePath = raw_input("Enter the capture file with path:")
```

```python
#read in pcap file and turn it into a packet list
packetList = rdpcap(capturePath)

print "Read of packet capture successful!"
```

```python
#create dictionary storing session information and packets
sessionOutput = packetList.sessions()

print "Session information pulled from packet capture!"
```

```python
#Dictionary to store information for session index translation
sessionMap = {}
```

```python
#counter used for session index
sessionCounter = 0

#for loop to save index number to sessionMap as key and Session info as data, also prints out the index and the session info
for k, v in sessionOutput.iteritems():
    sessionMap[sessionCounter] = k
    print sessionCounter
    print sessionMap[sessionCounter]
    print "\n"
    sessionCounter += 1
    
```

```python
print "Found " + str(len(sessionMap)) + " sessions in this packet capture!"
```
